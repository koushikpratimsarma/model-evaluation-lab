## 🤖 What is an AI Agent?

An AI Agent is an LLM that can:

✅ Reason

✅ Plan

✅ Use tools

✅ Remember

✅ Execute actions

Instead of only answering,

it decides

"What should I do next?"

## 📊 Agent Workflow

```
User Request

↓

Reason

↓

Plan

↓

Choose Tool

↓

Execute

↓

Observe

↓

Repeat (if needed)

↓

Final Answer
```

##  Model Context Protocol (MCP)

MCP provides a standard way for AI models to communicate with external tools and data sources.

Instead of writing custom integrations for every application,

one protocol can connect many tools.

Think of MCP as a common interface between an AI model and external systems.

##  Agent-to-Agent (A2A)

Sometimes one agent is not enough.

An agent can ask another agent for help.

Example

Travel Agent

↓

Weather Agent

↓

Hotel Booking Agent

↓

Flight Agent

They cooperate to complete a complex task.

````python
def search_documents(query):
    return f"Searching documents for: {query}"

def calculator(a, b):
    return a + b

question = "What is 15 + 20?"

if "+" in question:
    result = calculator(15, 20)
else:
    result = search_documents(question)

print(result)
````
