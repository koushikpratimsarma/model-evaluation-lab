# 09 - Sequence-to-Sequence (Seq2Seq)

---

In the previous notebook, we learned about GRU.

RNN, LSTM, and GRU can process sequential data.

But researchers wanted to solve a bigger problem:

How can one neural network convert an entire sequence into another sequence?

Examples:

English → French

Question → Answer

Speech → Text

This led to the development of the Encoder–Decoder (Seq2Seq) architecture.

## 📜 History

RNNs, LSTMs, and GRUs work well for many sequence tasks.

However, they produce one output for each input.

Researchers wanted a model that could:

- Read one complete sequence.
- Understand it.
- Generate another sequence.

For example:

English

↓

French

or

Question

↓

Answer

This became known as **Sequence-to-Sequence Learning (Seq2Seq)**.

## ❌ Why Previous Models Were Not Enough

Suppose we want to translate:

"I love artificial intelligence."

↓

French

"J'aime l'intelligence artificielle."

The input has four words.

The output also has four words.

Sometimes the output is longer.

Sometimes it is shorter.

Traditional RNNs were not designed for this flexible input-output mapping.

Researchers needed a new architecture.

## 💡 Think Like a Researcher

Imagine you work at Google in 2014.

Your LSTM understands sequences.

Now your manager asks:

"Translate English into French."

Question:

Should the network generate the translation while reading?

Or...

Should it first understand the whole sentence, then start translating?

Researchers chose the second idea.

This became the Encoder–Decoder architecture.

## 💡 Core Idea

A Seq2Seq model has two parts.

### Encoder

Reads the input sentence.

### Decoder

Generates the output sentence.

The Encoder passes its understanding of the sentence to the Decoder.

## 📊 Encoder–Decoder Architecture

```
English Sentence

↓

Encoder

↓

Context Vector

↓

Decoder

↓

French Sentence
```

The **Context Vector** contains the Encoder's understanding of the input.

````python
english = ["I", "love", "AI"]

french = ["J'", "aime", "IA"]

print("English:", english)
print("French :", french)
````

## 🧠 Human Analogy

Suppose someone says:

"I love artificial intelligence."

Before translating,

your brain first understands the sentence.

Only then do you speak the translation.

The Encoder works like understanding.

The Decoder works like speaking.

### Simulating Encoder

````python
sentence = "I love learning AI"

print("Input Sentence:")
print(sentence)

context_vector = "[0.82, -0.14, 0.56, 0.91]"

print("\nEncoder Output (Context Vector):")
print(context_vector)
````

### Simulating Decoder

````python
context_vector = "[0.82, -0.14, 0.56, 0.91]"

print("Received Context Vector:")
print(context_vector)

translation = "আমি AI শিখতে ভালোবাসি"

print("\nDecoder Output:")
print(translation)
````

````python
import torch
import torch.nn as nn

encoder = nn.LSTM(
    input_size=5,
    hidden_size=4,
    batch_first=True
)

print(encoder)
````

````python
# Encoder Output
x = torch.randn(1, 6, 5)

encoder_output, (hidden, cell) = encoder(x)

print("Encoder Output:", encoder_output.shape)

print("Hidden:", hidden.shape)

print("Cell:", cell.shape)
````

````python
# Decoder
decoder = nn.LSTM(
    input_size=5,
    hidden_size=4,
    batch_first=True
)

decoder_input = torch.randn(1, 1, 5)

decoder_output, _ = decoder(
    decoder_input,
    (hidden, cell)
)

print(decoder_output.shape)
````

## 🚨 The Context Vector Bottleneck

Everything the Encoder learns is compressed into **one Context Vector**.

```
Long Sentence

↓

Encoder

↓

One Vector

↓

Decoder
```

Question:

Can one vector remember an entire paragraph?

Usually...

No.

This became the biggest weakness of Seq2Seq.

````python
short_sentence = "I love AI"

long_sentence = """
I started learning programming many years ago,
then I studied machine learning,
deep learning,
computer vision,
natural language processing,
and finally built AI applications.
"""

print(short_sentence)

print()

print(long_sentence)

#Ask:
#Should both sentences be compressed into a single vector?
````

## ✅ Advantages

- Handles variable-length sequences
- Encoder–Decoder architecture
- Useful for translation
- Works with RNN, LSTM, and GRU

## ❌ Disadvantages

The Context Vector is a bottleneck.

Long sentences lose information.

The Decoder cannot easily access earlier words from the input.

Researchers asked:

"What if the Decoder could look back at the important words whenever it wants?"

That idea became **Attention**.

## 🌍 Applications

- Machine Translation
- Speech Recognition
- Text Summarization
- Dialogue Systems
- Question Answering
