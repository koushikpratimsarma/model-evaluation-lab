## 🧠 Transformer Usage

BERT

↓

Uses only the Encoder.

GPT

↓

Uses only the Decoder.

This is why they behave differently.

```python
#Encoder    
import torch
import torch.nn as nn

encoder_layer = nn.TransformerEncoderLayer(
    d_model=16,
    nhead=2,
    batch_first=True
)

encoder = nn.TransformerEncoder(
    encoder_layer,
    num_layers=2
)

x = torch.rand(1,5,16)

output = encoder(x)

print(output.shape)
```

```python
#Decoder
decoder_layer = nn.TransformerDecoderLayer(
    d_model=16,
    nhead=2,
    batch_first=True
)

decoder = nn.TransformerDecoder(
    decoder_layer,
    num_layers=2
)

memory = torch.rand(1,5,16)

target = torch.rand(1,5,16)

output = decoder(target,memory)

print(output.shape)
```
