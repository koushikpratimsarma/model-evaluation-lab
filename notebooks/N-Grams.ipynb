## 🧠 What is an N-gram?

An **N-gram** is a sequence of **N consecutive words (or characters)** from a piece of text.

The value of **N** determines how many words are grouped together.

---

## Types of N-grams

| N | Name | Example |
|---|------|---------|
| 1 | Unigram | AI |
| 2 | Bigram | I love |
| 3 | Trigram | I love AI |
| 4 | Four-gram | I love AI very |

---

## Example

Sentence:

```text
I love Machine Learning
```

### Unigrams (N = 1)

```text
I
love
Machine
Learning
```

---

### Bigrams (N = 2)

```text
I love
love Machine
Machine Learning
```

---

### Trigrams (N = 3)

```text
I love Machine
love Machine Learning
```

---

## Why do we use N-grams?

Unlike Bag of Words, which ignores word order, N-grams preserve some local sequence information.

Example:

Sentence 1:

```text
I love AI
```

Sentence 2:

```text
AI love I
```

### Bag of Words

Both sentences contain the same words, so they produce nearly the same representation.

### Bigrams

Sentence 1:

```text
I love
love AI
```

Sentence 2:

```text
AI love
love I
```

The bigrams are different, so the model can distinguish between the two sentences.

---

## Applications of N-grams

- Text Classification
- Sentiment Analysis
- Next Word Prediction
- Spell Checking
- Machine Translation
- Speech Recognition
- Search Engines
- Autocomplete

---

## Limitations

- Vocabulary grows rapidly as N increases.
- Many N-grams occur very rarely.
- Cannot capture long-distance dependencies.
- Large memory requirements for high values of N.

---

## Evolution

```text
Rule-Based NLP
      ↓
Bag of Words
      ↓
N-grams
      ↓
TF-IDF
      ↓
Word2Vec
      ↓
GloVe
      ↓
FastText
      ↓
RNN → LSTM → Transformer
```

N-grams improved over Bag of Words by preserving local word order, but they still struggle to understand long-range context and sentence meaning.

---

## 📌 Key Takeaway

An **N-gram** is a sequence of **N consecutive words or characters** used to preserve local word order in text. It provides more contextual information than Bag of Words but is limited in capturing long-range dependencies.
