# History

Before machine learning, computers could not learn from data.

Everything had to be written manually.

If we wanted a computer to detect sentiment, we wrote rules ourselves.

Example:

IF "good"
→ Positive

IF "bad"
→ Negative

These systems worked only for small problems.

As language became more complex, this approach became impossible to maintain.

This motivated researchers to look for machine learning methods.

## ❌ Limitation of Previous Approach

There is no previous model here.

Rule-Based System is the starting point of early NLP.

In this approach, humans manually write rules.

Example:

If sentence contains "good" → Positive  
If sentence contains "bad" → Negative

The computer does not learn from data.
It only follows human-written rules.

## 💡 Core Idea

Rule-Based NLP means solving language problems using manually written `if-else` rules.

The system checks whether certain words or patterns are present in the text.

For example:

- If text contains positive words → Positive
- If text contains negative words → Negative
- Otherwise → Unknown
