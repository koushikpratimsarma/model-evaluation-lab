## 📜 History

Word embeddings were a major breakthrough.

For the first time, computers could understand that words like:

- king
- queen
- man
- woman

are related.

However, researchers soon discovered another challenge.

A sentence is not just a collection of words.

The **order** of the words is also important.

This challenge became known as the **Sequence Problem**.

To solve it, researchers introduced Recurrent Neural Networks (RNNs).

## ❌ Why Word Embeddings Were Not Enough

Word2Vec, GloVe, and FastText represent **individual words**.

But language is made of **sentences**, not isolated words.

Consider these two sentences:

"I love AI"

"AI loves me"

Both contain similar words.

But the meanings are completely different.

Embeddings alone cannot understand this difference.

We need a model that understands **word order**.

## 💡 Core Idea

The meaning of a sentence depends on:

1. Which words appear
2. The order in which they appear

Changing the order can completely change the meaning.

This idea is called the **Sequence Problem**.

## 🧠 Another Challenge

Suppose we read a sentence one word at a time.

"The movie that I watched yesterday was amazing."

When reading the word:

"amazing"

the model should remember

"The movie"

If it forgets earlier words,

it cannot understand the sentence correctly.

This introduces another challenge:

Memory.

````python
sentence = "I love learning artificial intelligence"

words = sentence.split()

memory = []

for word in words:

    memory.append(word)

    print("Current Word :", word)

    print("Memory :", memory)

    print()
````

## 🧠 What is the Sequence Problem?

The **Sequence Problem** refers to the inability of early NLP models like **Bag of Words** and **TF-IDF** to understand the **order of words** in a sentence.

These models treat a sentence as an unordered collection (or "bag") of words, ignoring the sequence in which the words appear.

### Example 1

Sentence A:

```text
Dog bites man
```

Sentence B:

```text
Man bites dog
```

Bag of Words Representation:

```text
Dog : 1
Man : 1
Bites : 1
```

Both sentences produce the **same vector**, even though their meanings are completely different.

### Example 2

```text
I love AI
```

and

```text
AI love I
```

Bag of Words treats both sentences as identical because it only counts word occurrences.

---

## Why is this a problem?

The order of words often determines the meaning of a sentence.

Examples:

```text
I like Python.
```

```text
Python likes me.
```

or

```text
The cat chased the mouse.
```

```text
The mouse chased the cat.
```

Although they contain the same words, their meanings are completely different.

---

## Consequences

Ignoring word order leads to:

- Loss of sentence meaning
- Poor understanding of context
- Incorrect predictions in NLP tasks
- Inability to capture grammar and syntax

---

## Solution

To overcome the Sequence Problem, NLP evolved through models that preserve word order:

```text
Bag of Words
        ↓
Word2Vec
        ↓
RNN
        ↓
LSTM
        ↓
GRU
        ↓
Transformer
        ↓
BERT / GPT
```

Modern deep learning models process words sequentially (or using attention), allowing them to understand context, grammar, and relationships between words.

---

## 📌 Key Takeaway

The **Sequence Problem** is the limitation of early NLP methods where the order of words is ignored, causing sentences with different meanings but the same words to appear identical. Solving this problem led to the development of sequence models such as RNNs, LSTMs, and Transformers.
