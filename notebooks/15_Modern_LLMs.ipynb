## 💡 What is a Large Language Model?

A modern LLM is usually built from:

- Transformer Decoder
- Billions (or trillions) of tokens
- Millions or billions of parameters
- Massive GPU clusters

The architecture is still based on the Transformer.

The scale is what changed.

## 🧠 What are Parameters?

Parameters are the values the model learns during training.

Think of them as the model's learned knowledge.

## 📚 Context Window

A context window is the amount of text a model can consider at one time.

Small context:

One paragraph

Large context:

Many pages

Larger context windows help models reason over longer documents.

## 🧠 Pretraining

The first stage of training is called Pretraining.

The model learns from enormous amounts of text by predicting the next token.

## 🎯 Fine-Tuning

After pretraining,

the model can be adapted for specific tasks.

Examples

Medical QA

Legal Documents

Programming

Finance

Customer Support

This process is called Fine-Tuning.

##  Instruction Tuning

Early GPT models completed text.

Modern chat models follow instructions.

Example

Summarize this article.

Translate this sentence.

Write Python code.

Instruction tuning teaches the model to respond helpfully to user requests.

##  Reinforcement Learning from Human Feedback (RLHF)

Humans compare multiple model responses.

Helpful responses receive higher rewards.

The model gradually learns to produce answers that people prefer.

This process helped create conversational assistants like ChatGPT.

##  Mixture of Experts (MoE)

Instead of activating the entire model for every token,

an MoE model activates only a subset of specialized expert networks.

Benefits:

- Faster inference
- Better scalability
- Lower computation per token

##  Open vs Closed Models

Open-weight Models

- Llama
- DeepSeek
- Qwen

Closed Models

- GPT
- Claude
- Gemini

Open-weight models allow developers to run or fine-tune them locally (subject to their licenses), while closed models are accessed primarily through hosted APIs or services.

## ❌ Remaining Problems

Even modern LLMs have limitations.

They may:

- Hallucinate facts
- Lack up-to-date knowledge
- Be unaware of private company documents
- Cannot automatically access your PDFs or databases

How can we solve this?

Instead of retraining the model...

Can we retrieve relevant information first?
