## 🧠 What is Regex (Regular Expression)?

**Regex (Regular Expression)** is a sequence of special characters used to **search, match, extract, or replace patterns in text**.

Instead of searching for exact words, Regex searches for **patterns**.

---

## Why do we use Regex in NLP?

Regex is commonly used during **text preprocessing** to clean and extract useful information from text.

Examples:

- Remove punctuation
- Remove special characters
- Extract email addresses
- Extract phone numbers
- Extract dates
- Find URLs
- Validate input formats

---

## Example 1: Find Numbers

Text:

```text
My age is 22.
```

Regex:

```python
r"\d+"
```

Output:

```text
22
```

---

## Example 2: Find Email Addresses

Text:

```text
Contact: abc@gmail.com
```

Regex:

```python
r"\S+@\S+"
```

Output:

```text
abc@gmail.com
```

---

## Example 3: Remove Punctuation

Text:

```text
Hello, AI!
```

Regex:

```python
re.sub(r"[^\w\s]", "", text)
```

Output:

```text
Hello AI
```

---

## Common Regex Symbols

| Symbol | Meaning | Example |
|---------|---------|---------|
| `.` | Any single character | `a.c` → abc, a1c |
| `\d` | Any digit (0–9) | `\d+` → 123 |
| `\w` | Letter, digit, or underscore | `\w+` → hello123 |
| `\s` | Whitespace | space, tab |
| `+` | One or more occurrences | `a+` → a, aa, aaa |
| `*` | Zero or more occurrences | `a*` |
| `?` | Zero or one occurrence | `colou?r` |
| `^` | Start of text | `^Hello` |
| `$` | End of text | `world$` |

---

## Python Example

```python
import re

text = "My phone number is 9876543210."

numbers = re.findall(r"\d+", text)

print(numbers)
```

Output:

```text
['9876543210']
```

## Regex in Traditional NLP

Regex was widely used in early NLP systems for:

- Rule-Based Chatbots
- Text Cleaning
- Information Extraction
- Data Validation
- Pattern Matching

However, Regex **cannot understand the meaning or context** of language. It only matches predefined patterns.

---

## 📌 Key Takeaway

**Regex (Regular Expression)** is a powerful pattern-matching technique used to search, extract, validate, and clean text. It is an essential preprocessing tool in Traditional NLP but cannot understand semantics or context.
